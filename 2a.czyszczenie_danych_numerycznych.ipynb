# Oczyszczanie danych na podstawie funkcjonalności biblioteki pandas

```python
# Drop rows where column 'num-of-doors' has null values
#df = auto_data_dc.dropna(subset=['num-of-doors'])
auto_data_p = auto_data.dropna(subset=['num-of-doors'])
auto_data_p
```

```python
#add additional mean and median columns for price and horsepower
auto_data_p['price_mean'] = auto_data_p['price'].fillna(13207)
auto_data_p['price_median'] = auto_data_p['price'].fillna(10295)
auto_data_p['horsepower_mean'] = auto_data_p['horsepower'].fillna(104)
auto_data_p['horsepower_median'] = auto_data_p['horsepower'].fillna(95)
auto_data_p
```

```python
# remove original columns: price and horsepower from df
del auto_data_p['horsepower']  
del auto_data_p['price']  
auto_data_p
```

```python
# fill nulls with below values
auto_data_p.fillna({
    'normalized-losses': 122.0,
    'bore': 3.33,
    'stroke': 3.29,
    'peak-rpm': 5200,
}, inplace=True)

auto_data_p

# Fill null values in 'Column1' with a specified value (e.g., 0)
# df['Column1'].fillna(value=0, inplace=True)
```

## Oczyszczanie danych z biblioteką Scikit-learn

4. Na podstawie utworzonej kopii ramki z punktu 2, powtórzyć kroki z punktu 3, ale przy użyciu biblioteki *Scikit-learn*.

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
```

```python
auto_data
```

Przed właściwym uzupełnieniem wartości należy najpierw wybrać zestaw pasujących atrybutów, dla których zostanie zastosowana wybrana strategia.

```python
#auto_data_dc[['normalized-losses','horsepower','peak-rpm','price']] = auto_data_dc[['normalized-losses','horsepower','peak-rpm','price']].dropna().astype(int)
# auto_data_dc['normalized-losses'] = auto_data_dc['normalized-losses'].dropna().astype(int)
# auto_data_dc['horsepower'] = auto_data_dc['horsepower'].dropna().astype(int)
# auto_data_dc['peak-rpm'] = auto_data_dc['peak-rpm'].dropna().astype(int)
# auto_data_dc['price'] = auto_data_dc['price'].dropna().astype(int)
# auto_data_dc['stroke'] = auto_data_dc['stroke'].dropna().astype(float)
# auto_data_dc['bore'] = auto_data_dc['bore'].dropna().astype(float)
num_attributes = auto_data.select_dtypes(include=['number'])
```

```python
num_attributes
```

```python
num_attributes.isnull().any(axis=0)
```

Wywołanie metody *fit* na utworzonym obiekcie pozwoli na automatyczne wyznaczenie wartości do uzupełnienia w każdym z atrybutów.

```python
imputer.fit(num_attributes)
```

W atrybucie *statistics_* zawarte są wyznaczone wartości do zastąpenia brakujących według obranej strategii.

```python
imputer.statistics_
```

Do zastąpienia brakujących wartości przeznaczona jest metoda *transform*.

```python
new_num_attributes = imputer.transform(num_attributes)
new_num_attributes
```

```python
import pandas as pd

new_num_attributes = pd.DataFrame(new_num_attributes, columns=num_attributes.columns)
```
