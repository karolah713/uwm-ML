1. Korzystając ze zbioru danych `mnist` sprawdzić interfejsy następujących klasyfikatorów:
  - Regresja Logistyczna: [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression)
  - Liniowa Maszyna Wektorów Nośnych: [LinearSVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVC.html#sklearn.svm.LinearSVC).
  
  Porównać wyniki dokładności globalnej i zbalansowanej.


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, balanced_accuracy_score
pd.read_csv
```

```python
from sklearn.datasets import fetch_openml
```

```python
# podział na cechy X i atrybut decyzyjny y
mnist_X, mnist_y = fetch_openml('mnist_784', return_X_y=True, as_frame=True)
mnist_X
```

```python
mnist_y
```

```python
# podział na zbiór treningowy i testowy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mnist_X, mnist_y, test_size=.1, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(max_iter=2000)
clf.fit(X_train_scaled, y_train)
```

```python
y_pred = clf.predict(X_test_scaled)
```

```python
accuracy_score(y_test, y_pred)
```

```python
balanced_accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
```

```python
clf2 = make_pipeline(
    MinMaxScaler(),
    LinearSVC(dual="auto", random_state=0, tol=1e-5))
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred = clf2.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
balanced_accuracy_score(y_test, y_pred)
```

2. Korzystając ze zbioru danych `CaliforniaHousing` sprawdzić interfejsy następujących regresorów:
  - Regresja Bayesowska: [BayesianRidge](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.BayesianRidge.html#sklearn.linear_model.BayesianRidge)
  - Regresor Wektorów Nośnych: [SVR](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html#sklearn.svm.SVR).
  
  Porównać wyniki błędu średniokwadratowego i jego pierwiastka oraz średniego błędu bezwzględnego.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
from sklearn.datasets import fetch_california_housing
all_data = fetch_california_housing(as_frame=True)['frame']
house_X = fetch_california_housing(as_frame=True)['data']   #cechy 
house_y = fetch_california_housing(as_frame=True)['target'] #atrybut deczyjny
house_X
```

```python
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(house_X, house_y, test_size=0.1, random_state=42)
```

```python
X_train_h.shape, X_test_h.shape, y_train_h.shape, y_test_h.shape
```

```python
from sklearn.linear_model import BayesianRidge
```

```python
clf3 = make_pipeline(
    MinMaxScaler(),
    BayesianRidge())
```

```python
clf3.fit(X_train_h, y_train_h)
```

```python
y_pred = clf3.predict(X_test_h)
y_pred.shape
```

```python
mse = mean_squared_error(y_test_h, y_pred)
mae = mean_absolute_error(y_test_h, y_pred)
```

```python
f'MSE: {mse}, RMSE: {sqrt(mse)}, MAE: {mae}'
```

```python
from sklearn.svm import SVR
```

```python
clf4 = make_pipeline(
    MinMaxScaler(),
    SVR(C=1.0, epsilon=0.2))
```

```python
clf4.fit(X_train_h, y_train_h)
```

```python
y_pred = clf4.predict(X_test_h)
```

```python
mse = mean_squared_error(y_test_h, y_pred)
mae = mean_absolute_error(y_test_h, y_pred)
```

```python
f'MSE: {mse}, RMSE: {sqrt(mse)}, MAE: {mae}'
```
